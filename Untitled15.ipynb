```python
import math


def show_sample_images(base_path, class_names):
    num_classes = len(class_names)
    cols = 4
    rows = math.ceil(num_classes / cols)

    plt.figure(figsize=(15, rows * 3))

    for idx, class_name in enumerate(class_names):
        class_folder = os.path.join(base_path, class_name)
        image_files = [f for f in os.listdir(class_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

        if image_files:
            img_path = os.path.join(class_folder, image_files[0])
            img = mpimg.imread(img_path)

            plt.subplot(rows, cols, idx + 1)
            plt.imshow(img)
            plt.title(class_name, fontsize=8)
            plt.axis('off')

    plt.tight_layout()
    plt.show()

    #call the function
show_sample_images(dataset_path, class_folders)

```

```python
dataset_path = '/content/drive/MyDrive/balanced_dataset/balanced_dataset'
img_size = 224
```

```python
# Select 15 Common Skin Conditions
selected_classes = [
    'Acne And Rosacea Photos',
    'Actinic Keratosis Basal Cell Carcinoma And Other Malignant Lesions',
    'Atopic Dermatitis Photos',
    'Cellulitis Impetigo And Other Bacterial Infections',
    'Eczema Photos',
    'Herpes Hpv And Other Stds Photos',
    'Nail Fungus And Other Nail Disease',
    'Scabies Lyme Disease And Other Infestations And Bites',
    'Seborrheic Keratoses And Other Benign Tumors',
    'Systemic Disease',
    'Tinea Ringworm Candidiasis And Other Fungal Infections',
    'Urticaria Hives',
    'Vi Chickenpox',
    'Warts Molluscum And Other Viral Infections'
]
images, labels = [], []
```

```python
# Load and Process Images
images, labels = [], []

for idx, class_name in enumerate(selected_classes):
    folder_path = os.path.join(dataset_path, class_name)
    if not os.path.exists(folder_path):
        print(f"Skipping missing class: {class_name}")
        continue

    image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

    for img_file in image_files:
        try:
            img_path = os.path.join(folder_path, img_file)
            img = Image.open(img_path).convert('L')  # Grayscale
            img = img.resize((img_size, img_size))
            img_array = np.array(img) / 255.0
            images.append(img_array)
            labels.append(idx)
        except:
            continue
```

```python
#Filter out underrepresented classes (<2 images)
label_counts = Counter(labels)
X_filtered, y_filtered = [], []
for img, lbl in zip(images, labels):
    if label_counts[lbl] >= 2:
        X_filtered.append(img)
        y_filtered.append(lbl)

X = np.array(X_filtered).reshape(-1, img_size, img_size, 1)
X = np.repeat(X, 3, axis=-1)
y = to_categorical(np.array(y_filtered), num_classes=len(selected_classes))
```

```python
#Prepare Dataset
X = np.array(images).reshape(-1, img_size, img_size, 1)
y = to_categorical(np.array(labels), num_classes=len(selected_classes))
```

```python

# Import Counter from collections
from collections import Counter
print("Class counts:", dict(Counter(labels)))
```

```python
#4. Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
#5. Convert Grayscale to 3 Channels
X_train = np.repeat(X_train, 3, axis=-1)
X_test = np.repeat(X_test, 3, axis=-1)
```

```python
#Data Augmentation
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)
datagen.fit(X_train)
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam

# Define base model
base_model = MobileNetV2(include_top=False, weights='imagenet', input_tensor=Input(shape=(img_size, img_size, 3)))

# UNFREEZE LAST 20 LAYERS FOR FINE-TUNING
for layer in base_model.layers[:-20]:
    layer.trainable = False
for layer in base_model.layers[-20:]:
    layer.trainable = True

#BUILD THE HEAD
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)  # prevent overfitting
output = Dense(len(selected_classes), activation='softmax')(x)

#COMPILE FINAL MODEL9
model = Model(inputs=base_model.input, outputs=output)
model.compile(optimizer=Adam(learning_rate=1e-4), loss='categorical_crossentropy', metrics=['accuracy'])

```

```python
#8. Train Model
history = model.fit(datagen.flow(X_train, y_train, batch_size=32),
                    validation_data=(X_test, y_test),
                    epochs=5)
```

```python
#  Evaluate
loss, acc = model.evaluate(X_test, y_test)
print(f"\n Test Accuracy: {acc:.4f}")
```
